```python
import  numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv, det
from distance import bhatt_dis_gen, bhatt_dis, maha_dis
import pickle
import corner
```

```python
data=pickle.load(open('data.pkl', 'rb'))
data_pure=pickle.load(open('data_pure.pkl', 'rb'))
```

```python
A1,gamma1,A2,gamma2=[data[i][0] for i in range(len(data))]
a1,g1,a2,g2=[data_pure[i][0] for i in range(len(data_pure))]
```

```python
np.mean(A1)-np.mean(A2), np.mean(gamma1)-np.mean(gamma2)
```

# Pure Gaussian Data 

```python
mean=np.array([[-14,3],[-14.0001,3.00001]])
cov1 = np.array([[2, 1], [1, 3]])
cov2 = np.array([[3, 1.5], [1.5, 1]])
```

```python
bhatt_dis(mean[0],mean[1],cov1,cov2)
```

```python
bhatt_dis(np.array([np.mean([a1]),np.mean(g1)]),np.array([np.mean([a2]),np.mean(
    g2)]),np.cov([a1,g1]),np.cov([a2,g2]))
```

```python
bhatt_dis_gen(a1,g1,a2,g2)
```

# Noisy Gaussian Data

```python
bhatt_dis_gen(A1,gamma1,A2,gamma2)
```

```python
bhatt_dis(np.array([np.mean([A1]),np.mean(gamma1)]),np.array([np.mean([A2]),np.mean(
    gamma2)]),np.cov([A1,gamma1]),np.cov([A2,gamma2]))
```

```python
maha_dis(A1,gamma1,A2,gamma2)
```

```python
maha_dis(a1,g1,a2,g2)
```

```python
bhatt_dis_gen(A1,gamma1,A2,gamma2)/maha_dis(A1,gamma1,A2,gamma2)
```
